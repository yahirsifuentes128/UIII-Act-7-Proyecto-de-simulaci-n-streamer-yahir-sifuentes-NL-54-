```python
# =========================================
# STREAMER Y ESPECTADOR - IA SIMPLE
# CBTIS 128
# Alumno: Yahir Sifuentes Vargas
# =========================================

print("===================================")
print("   IA SIMPLE PARA STREAMERS")
print("===================================")

nombre = input("Ingresa tu nombre: ")

seguir = "si"

while seguir == "si":

    print("\nHola", nombre)
    print("\nMENU PRINCIPAL")
    print("1. Entrar como Streamer")
    print("2. Entrar como Espectador")
    print("3. Ver nivel de fama")
    print("4. Salir")

    opcion = input("Selecciona una opcion: ")

    # STREAMER
    if opcion == "1":

        viewers = int(input("¿Cuantos espectadores tienes?: "))

        if viewers >= 1000:
            print("🔥 Eres un streamer SUPER FAMOSO")
            print("La IA recomienda hacer un torneo en vivo")

        elif viewers >= 500:
            print("😎 Tu canal esta creciendo rapido")
            print("La IA recomienda subir clips a TikTok")

        elif viewers >= 100:
            print("🙂 Vas empezando pero tienes futuro")
            print("La IA recomienda interactuar con el chat")

        else:
            print("😅 Canal pequeño")
            print("La IA recomienda hacer directos mas seguido")

    # ESPECTADOR
    elif opcion == "2":

        horas = int(input("¿Cuantas horas ves streams al dia?: "))

        if horas >= 8:
            print("📺 Eres un espectador EXTREMO")
            print("La IA recomienda descansar un poco")

        elif horas >= 4:
            print("🎮 Te gustan mucho los streams")
            print("La IA recomienda seguir nuevos creadores")

        elif horas >= 1:
            print("👍 Eres espectador casual")
            print("La IA recomienda apoyar con un like")

        else:
            print("😴 Casi no ves streams")
            print("La IA recomienda explorar Twitch o YouTube")

    # NIVEL DE FAMA
    elif opcion == "3":

        seguidores = int(input("Ingresa tus seguidores: "))

        if seguidores >= 100000:
            print("🏆 Nivel LEYENDA DE INTERNET")

        elif seguidores >= 10000:
            print("⭐ Nivel STREAMER POPULAR")

        elif seguidores >= 1000:
            print("✨ Nivel CREADOR EN CRECIMIENTO")

        else:
            print("🌱 Nivel NOVATO")

    # SALIR
    elif opcion == "4":

        print("Gracias por usar la IA para streamers")
        seguir = "no"

    # OPCION INVALIDA
    else:
        print("❌ Opcion no valida")

print("Programa realizado por yahir sifuentes ")
```
