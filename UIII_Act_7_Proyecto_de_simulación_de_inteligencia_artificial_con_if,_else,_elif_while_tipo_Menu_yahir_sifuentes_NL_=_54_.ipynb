```python
# --- SIMULADOR DE ASISTENTE IA PARA STREAMERS ---

# 1. Entrada de datos interactiva
nombre_streamer = input("Nombre del Streamer: ")
viewers = int(input("Cantidad de espectadores actuales: "))
subs_actuales = int(input("Cantidad de suscriptores actuales: "))
mensaje_chat = input("Último mensaje recibido en el chat: ")
es_moderador = input("¿El autor es moderador? (si/no): ") == "si"

# 2. Definición de metas (Operadores aritméticos)
meta_subs = 100
subs_faltantes = meta_subs - subs_actuales  # Resta aritmética [3]

# 3. Lógica de la "IA" (Estructuras condicionales y Operadores)

print(f"\n--- Reporte de IA para {nombre_streamer} ---")

# Caso 1: Detección de mensajes virales o alertas (Operadores Lógicos y Relacionales)
# Verificamos si hay muchos espectadores Y el mensaje es muy largo (simulando spam)
if viewers > 5000 and len(mensaje_chat) > 50:  # Operador 'and' y '>' [4, 5]
    print("ALERTA IA: Posible ataque de SPAM detectado debido al alto volumen de tráfico.")
    print("ACCIÓN: Activando modo lento automáticamente.")

# Caso 2: Interacción con suscriptores y metas (Operadores Relacionales y Lógicos)
elif subs_actuales >= meta_subs or viewers > 10000: # Operador 'or' [6]
    print("ESTADO: ¡Meta de comunidad alcanzada o stream en tendencia!")
    print(f"Felicidades {nombre_streamer}, el stream es un éxito total.")

# Caso 3: Seguridad del Chat (Operador Lógico 'not' y de Identidad)
# Si el mensaje contiene una palabra prohibida y NO es un moderador
elif ("insulto" in mensaje_chat.lower()) and not es_moderador: # Operador 'not' [6]
    print("ALERTA IA: Lenguaje inapropiado detectado de un usuario común.")
    print("ACCIÓN: Mensaje eliminado y usuario advertido.")

# Caso 4: Análisis de interacción baja (Operadores Relacionales y Aritméticos)
elif viewers < 50 and subs_faltantes > 50: # Uso de comparación y variable calculada [4]
    print("SUGERENCIA IA: La interacción es baja.")
    print(f"Te faltan {subs_faltantes} subs para la meta. ¡Prueba hacer un sorteo!")

# Caso por defecto (Estructura else)
else:
    print("ESTADO: El stream transcurre con normalidad.")
    print("IA: Monitoreando interacciones silenciosamente...")

print("programa realizado por ayhir sifuentes NL = 54 ")
```
